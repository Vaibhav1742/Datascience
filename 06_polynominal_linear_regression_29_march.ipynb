```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd 
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

```python
df =pd.read_csv(R"C:\Users\raval\PycharmProjects\MachineLearning\polynominal_data_29_march_orders.csv")
plt.scatter(df. slot , df. amount)
# LinearRegression Polynon1alFeatures
poly = PolynomialFeatures(degree=2)
x_poly = poly.fit_transform(df[["slot"]])
plt.show()
```

```python
# # this shows linear regr
# reg = LinearRegression()
# reg.fit(df[["slot"]],df[["amount"]])
```

for ploting curve line for prediction --> polynominal regression

```python
reg = LinearRegression()
reg.fit(x_poly,df[["amount"]])
plt.plot(df[["slot"]].values,reg.predict(x_poly))
```

<div class="alert alert-info"><font size=5 color="blue">By incresing the value of degree -- > more efficient curve is generated (accurate prediction)</font></div>

```python
plt.scatter(df. slot , df. amount)
poly = PolynomialFeatures(degree=2)
x_poly = poly.fit_transform(df[["slot"]])
reg = LinearRegression()
reg.fit(x_poly,df[["amount"]])
plt.plot(df[["slot"]].values,reg.predict(x_poly))
plt.show()
```
